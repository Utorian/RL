```python
import gymnasium as gym
from gymnasium.wrappers import TimeLimit
```

## Declaring the environment
This script is working out of the box with environment from Toy Text

We are using **render_mode = ansi**, which means that we need to print the render function to see the current status of the enviroment. 
If you would like to use **render_mode=human** instead you need to delete the print() command.


```python
env = gym.make("FrozenLake-v1", render_mode="ansi")
observation, info = env.reset()
```

```python
print(env.render())
```

Declaring how many games we would like the agent to play in the variable: **how_many_games_to_play**.
In **max_action_pr_game** variable we set how many action the agent can make before quiting this game iteration

```python
max_action_pr_game = 10
how_many_games_to_play = 100
```

Using the TimeLimit Wrapper to activate the new parameter truncate from the step(). This will be activated when max_action_pr_game occour
https://gymnasium.farama.org/api/wrappers/misc_wrappers/

```python
env = TimeLimit(env, max_episode_steps=max_action_pr_game)
```

```python
for i in range(how_many_games_to_play):
    env.reset()    
    
    done = False    
    truncate = False
    print("**************** New Game *********************")
    while not done:         
        next_state,reward,done,truncate,info = env.step(env.action_space.sample())
        print(env.render())
        if truncate:
            break
    
```
