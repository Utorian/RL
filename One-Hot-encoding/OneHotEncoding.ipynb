One-Hot-Encoding using pandas

```python
import pandas as pd
```

```python
color_data = pd.read_csv('data.csv')
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(handle_unknown='ignore')
```

```python
encoded_data = pd.DataFrame(encoder.fit_transform(color_data[['COLOR']]).toarray())
```

```python
encoded_data
```

```python
color_data = color_data.join(encoded_data)
```
